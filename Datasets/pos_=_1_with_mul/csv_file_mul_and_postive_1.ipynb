## CSV to list

```python
import csv

with open('INPUT_Pos.csv', 'r') as f:
  reader = csv.reader(f)
  my_list = list(reader)
```

## Multiply Multiply by a certain number

```python
import numpy as np

new_list = list(np.float_(my_list))


my_new_list = []
for i in new_list:
    my_new_list.append(i * 1.0005)
```

```python
print(len(my_new_list))
```

## Save the CSV file

```python
import numpy as np
myarray = np.asarray(my_new_list)
np.savetxt("INPUT_Pos_mul.csv", myarray, delimiter=",")
```

## Load new data after multiplication

```python
fileName = 'INPUT_Pos_mul.csv'
dataset = pd.read_csv(fileName, header=None, engine='python')
dataset.shape
```

```python
dataset.head(5)
```

## Positive = 1

```python
dummy_one = np.ones(len(dataset), dtype='int')
series = pd.Series(dummy_one)
dataset = pd.concat([dataset, series], axis=1)
```

## After positive = 1 and save the new CSV file

```python
dataset.to_csv('INPUT_Positive.csv')
```
