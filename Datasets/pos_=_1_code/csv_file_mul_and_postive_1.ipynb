# CSV file 1

## Load dataset

```python
import pandas as pd

fileName = 'Meth_pssm+spd_positive.csv'
dataset = pd.read_csv(fileName, header=None, engine='python')
dataset.shape
```

```python
dataset.head(5)
```

## Positive = 1

```python
import numpy as np

dummy_one = np.ones(len(dataset), dtype='int')
series = pd.Series(dummy_one)
dataset = pd.concat([dataset, series], axis=1)
```

```python
dataset.shape
```

```python
dataset.head(5)
```

## After positive = 1 and save the new CSV file

```python
dataset.to_csv('pssm+spd_Pos.csv')
```
